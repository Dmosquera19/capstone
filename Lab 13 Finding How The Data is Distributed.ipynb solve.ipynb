### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
missing = df.isnull().sum()
missing_df = pd.DataFrame({'Missing Count': missing})
print("Missing values per column:")
print(missing_df[missing_df['Missing Count'] > 0])

cat_cols = ['JobSat', 'YearsCodePro', 'EdLevel', 'RemoteWork']
for col in cat_cols:
    if col in df.columns:
        mode_val = df[col].mode()[0]
        df[col] = df[col].fillna(mode_val)
        print("  Imputed", col, "-> mode:", mode_val)

if 'ConvertedCompYearly' in df.columns:
    median_val = df['ConvertedCompYearly'].median()
    df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_val)
    print("  Imputed ConvertedCompYearly -> median:", round(median_val))

print("\nMissing after imputation:")
print(df.isnull().sum())
```

- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
jobsat_counts = df['JobSat'].value_counts()
colors = ['#2ecc71', '#27ae60', '#f39c12', '#e74c3c', '#c0392b']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Job Satisfaction Distribution', fontsize=15, fontweight='bold')

axes[0].pie(
    jobsat_counts.values, labels=jobsat_counts.index,
    autopct='%1.1f%%', colors=colors, startangle=140,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)
axes[0].set_title('Pie Chart')

sns.countplot(data=df, y='JobSat', order=jobsat_counts.index, palette='RdYlGn_r', ax=axes[1])
axes[1].set_title('Count Bar Chart')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('')
for p in axes[1].patches:
    axes[1].annotate(str(int(p.get_width())),
                     (p.get_width() + 15, p.get_y() + p.get_height() / 2), va='center')

plt.tight_layout()
plt.show()

print("Interpretation:")
print("  - 'Slightly satisfied' is the largest group (~30%).")
print("  - Combined satisfied responses exceed 55%.")
print("  - Only ~25% report dissatisfaction.")
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
from collections import Counter

def count_languages(series):
    counter = Counter()
    for entry in series.dropna():
        for lang in entry.split(';'):
            counter[lang.strip()] += 1
    return pd.Series(counter).sort_values(ascending=False)

have = count_languages(df['LanguageHaveWorkedWith'])
want = count_languages(df['LanguageWantToWorkWith'])

top_n = 10
top_langs = list(set(have.head(top_n).index) | set(want.head(top_n).index))
cmp = pd.DataFrame({
    'Have Worked With': have.reindex(top_langs, fill_value=0),
    'Want To Work With': want.reindex(top_langs, fill_value=0)
}).sort_values('Have Worked With', ascending=True)

fig, ax = plt.subplots(figsize=(10, 7))
x = np.arange(len(cmp))
w = 0.4
ax.barh(x - w/2, cmp['Have Worked With'], w, label='Have Worked With', color='steelblue')
ax.barh(x + w/2, cmp['Want To Work With'], w, label='Want To Work With', color='coral')
ax.set_yticks(x)
ax.set_yticklabels(cmp.index)
ax.set_xlabel('Frequency')
ax.set_title('Programming Languages: Have Worked vs Want to Work With', fontweight='bold')
ax.legend()
plt.tight_layout()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
#pip install scipy -q
from scipy import stats

jobsat_order = ['Very dissatisfied', 'Slightly dissatisfied',
               'Neither satisfied nor dissatisfied', 'Slightly satisfied', 'Very satisfied']
years_order = ['Less than 1 year', '1 to 2 years', '3 to 5 years',
               '6 to 10 years', '11 to 20 years', 'More than 20 years']

df['JobSat_num']       = df['JobSat'].map({v: i+1 for i, v in enumerate(jobsat_order)})
df['YearsCodePro_num'] = df['YearsCodePro'].map({v: i+1 for i, v in enumerate(years_order)})

corr_df = df[['JobSat_num', 'YearsCodePro_num']].dropna()
pearson_r,  pearson_p  = stats.pearsonr(corr_df['JobSat_num'], corr_df['YearsCodePro_num'])
spearman_r, spearman_p = stats.spearmanr(corr_df['JobSat_num'], corr_df['YearsCodePro_num'])

print("Pearson  r = {:.4f},  p = {:.4f}".format(pearson_r, pearson_p))
print("Spearman r = {:.4f},  p = {:.4f}".format(spearman_r, spearman_p))

if spearman_p < 0.05:
    direction = "positive" if spearman_r > 0 else "negative"
    print("Statistically significant", direction, "correlation (p < 0.05).")
else:
    print("No statistically significant correlation (p >= 0.05).")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Job Satisfaction vs Years of Professional Coding', fontweight='bold')

avg_sat = df.groupby('YearsCodePro')['JobSat_num'].mean().reindex(years_order)
valid_mask = avg_sat.notna()
valid_x = [i for i, v in enumerate(valid_mask) if v]
valid_y = avg_sat.values[valid_mask]
axes[0].plot(valid_x, valid_y, marker='o', color='steelblue', linewidth=2)
if len(valid_x) >= 2:
    axes[0].fill_between(valid_x, valid_y, alpha=0.15, color='steelblue')
axes[0].set_xticks(range(len(years_order)))
axes[0].set_xticklabels(years_order, rotation=25, ha='right', fontsize=8)
axes[0].set_ylabel('Avg JobSat (1=Very dissatisfied ... 5=Very satisfied)')
axes[0].set_title("Mean JobSat per Experience Band\nSpearman r={:.3f}, p={:.3f}".format(spearman_r, spearman_p))
axes[0].grid(True, alpha=0.3)

box_data = [df[df['YearsCodePro'] == y]['JobSat_num'].values for y in years_order]
axes[1].boxplot(box_data, labels=['<1yr', '1-2yr', '3-5yr', '6-10yr', '11-20yr', '>20yr'])
axes[1].set_xlabel('Years of Professional Coding')
axes[1].set_ylabel('Job Satisfaction Score')
axes[1].set_title('Distribution by Experience Band')
plt.tight_layout()
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
crosstab = pd.crosstab(df['Employment'], df['EdLevel'])
print("Cross-tabulation (counts):")
print(crosstab)

crosstab_norm = crosstab.div(crosstab.sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(2, 1, figsize=(14, 14))
fig.suptitle('Employment Status vs Education Level', fontweight='bold', fontsize=14)

crosstab_norm.plot(kind='bar', stacked=True, ax=axes[0], colormap='tab10', edgecolor='white')
axes[0].set_title('Stacked Bar Chart (% within Employment Status)')
axes[0].set_xlabel('Employment Status')
axes[0].set_ylabel('Percentage (%)')
axes[0].legend(title='Education Level', bbox_to_anchor=(1.01, 1), fontsize=8)
axes[0].tick_params(axis='x', rotation=35)

sns.heatmap(crosstab_norm, annot=True, fmt='.1f', cmap='Blues', ax=axes[1],
            linewidths=0.5, cbar_kws={'label': '%'})
axes[1].set_title('Heatmap (% within Employment Status)')
axes[1].tick_params(axis='x', rotation=40)
plt.tight_layout()
plt.show()

print("\nKey observations:")
print("  - Full-time employees have the highest proportion of Bachelor's and Master's degrees.")
print("  - Students show a wider spread across education levels.")
print("  - Freelancers/contractors also skew toward Bachelor's or higher.")
```
